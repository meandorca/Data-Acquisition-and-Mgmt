```python
import pandas as pd
cumulative_miles = pd.Series([55, 120, 190, 260],
                             index=["Day 1", "Day 2", "Day 3", "Day 4"])
daily_miles = cumulative_miles.diff()
daily_miles.iloc[0] = cumulative_miles.iloc[0]
print("Cumulative miles:")
display(cumulative_miles)
print("\nMiles ridden each day:")
display(daily_miles)
```
