```python
import pandas as pd

df = pd.read_csv("movieratings.csv")
df = df.set_index("User")
df
```

```python
movie_avg = df.mean(axis=0)

print("Average rating for each movie:")
print(movie_avg)
```

```python
normalized_df = df.sub(df.mean(axis=1), axis=0)

normalized_df
```

```python
normalized_df.mean(axis=1)
normalized_df.mean(axis=0)
```

```python
standardized_df = df.sub(df.mean(axis=1), axis=0).div(df.std(axis=1), axis=0)

standardized_df
```

```python
standardized_df.mean(axis=1)
standardized_df.mean(axis=0)
```

###Conclusion

  Normalized ratings help remove personal bias from users who tend to rate movies higher or lower than others. By subtracting each user’s average rating, we can compare movies based on how much a person liked them relative to their usual rating habits rather than just the raw score. This makes it easier to see which movies stood out for people even if they normally rate things differently.
  
  One advantage of normalized ratings is that they reduce user bias and make comparisons between different users more consistent. This is especially useful in systems like recommendation engines because it helps highlight which movies people liked more than their typical ratings.
  
  However, there are also some disadvantages. Normalization removes the original 1–5 rating scale, which can make the results harder to interpret at first. A movie that was rated highly by everyone might not appear as strong after normalization because the values are based on relative preference rather than the actual scores. Normalization can also exaggerate small differences when only a few ratings are available.
  
  Standardized ratings go one step further by also accounting for how much each person’s ratings vary. This can make comparisons even more balanced, but with a small dataset it can sometimes introduce extra noise.
