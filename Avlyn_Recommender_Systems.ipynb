# Analysis of YouTube’s Recommendation System Using Animal Content

## Introduction
This report analyzes how YouTube recommends animal related content like dog videos and wildlife clips. The goal is to understand how the system works, evaluate the user experience, and suggest improvements.

## Scenario Design – User Perspective

**1) Who are the target users?**
Users of YouTube who watch animal related content such as dog and wildlife videos.

**2) What are their key goals?**
To find enjoyable animal videos quickly and keep watching content they like.

**3) How can the system help them accomplish those goals?**
YouTube recommends videos based on what users watch, like, and click on so they can easily find similar animal content.

## Scenario Design – Company Perspective

**1) Who are the target users?**
All YouTube users, including those watching animal content.

**2) What are their key goals?**
To keep users watching videos and increase engagement.

**3) How can the system help them accomplish those goals?**
YouTube uses recommendation systems to suggest videos that users are likely to watch, helping increase watch time.

## Reverse Engineering the Recommendation System

Based on observation, YouTube seems to use several techniques:

- Collaborative filtering: recommending videos watched by similar users
- Content based filtering: recommending videos with similar tags (like dogs, pets, wildlife)
- Engagement based ranking: prioritizing videos with high watch time and interaction

For example, after watching multiple dog videos, the homepage becomes filled with similar content.

## Recommendations for Improvement

- Increase content diversity: avoid showing only one type of animal (only dogs)
- Reduce repetition: limit recommending very similar videos repeatedly
- Improve user controls: allow users to refine their preferences more easily
- Adapt faster to changing interests: if a user switches from pets to wildlife, recommendations should obviously adjust quickly

## Conclusion

YouTube’s recommendation system is actually very effective at personalizing content but it can show too much of the same thing. By showing more variety and letting users control what they see, YouTube could make watching better for everyone.
